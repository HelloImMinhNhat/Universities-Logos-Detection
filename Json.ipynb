```python
import os
import cv2
import numpy as np
from pycocotools.coco import COCO

# Đường dẫn đến tệp annotations COCO JSON
annotations_file = './mask_rbf/valid/_annotations.coco.json'
# Thư mục để lưu các hình ảnh mask
output_dir = './Mask/valid'

# Tạo thư mục nếu nó chưa tồn tại
os.makedirs(output_dir, exist_ok=True)

# Tạo đối tượng COCO từ tệp annotations
coco = COCO(annotations_file)

# Lặp qua mỗi danh mục
for cat_id in coco.getCatIds():
    category_info = coco.loadCats(cat_id)[0]
    category_name = category_info['name']

    # Tạo thư mục tương ứng cho danh mục
    category_output_dir = os.path.join(output_dir, category_name)
    os.makedirs(category_output_dir, exist_ok=True)

    # Lặp qua mỗi hình ảnh trong tập tin COCO
    for img_id in coco.catToImgs[cat_id]:
        img_info = coco.loadImgs(img_id)[0]

        # Lấy các chú thích thuộc danh mục hiện tại cho hình ảnh
        ann_ids = coco.getAnnIds(imgIds=img_id, catIds=cat_id)
        anns = coco.loadAnns(ann_ids)

        # Tạo mask cho các chú thích của danh mục hiện tại
        result = np.zeros((img_info['height'], img_info['width']), dtype=np.uint8)
        for ann in anns:
            mask = coco.annToMask(ann)
            cls_id = ann['category_id']
            mask *= cls_id
            result += mask

        # Chuyển các giá trị pixel vượt quá số lượng lớp mong muốn thành 0
        num_classes = len(coco.getCatIds())
        result[result > num_classes] = 0

        # Chuyển các giá trị pixel khác 0 thành 255 (màu trắng)
        result[result != 0] = 255

        # Lưu hình ảnh mask với tên tương ứng vào thư mục của danh mục
        file_name_parts = img_info['file_name'].split('_')
        file_name_out = file_name_parts[1].split('.')
        new_file_name = f"{file_name_parts[0]}_{file_name_out[0]}.jpg"

        # Lưu hình ảnh mask với tên tương ứng vào thư mục của danh mục
        output_path = os.path.join(category_output_dir, new_file_name)
        cv2.imwrite(output_path, result)
```

```python
import os
import re

def rename_files_in_folder(folder_path):
    # Kiểm tra xem đường dẫn thư mục có tồn tại không
    if not os.path.exists(folder_path):
        print(f"Thư mục '{folder_path}' không tồn tại.")
        return

    # Lấy danh sách tất cả các tệp trong thư mục
    files = os.listdir(folder_path)

    # Lặp qua từng tệp và đổi tên
    for file_name in files:
        # Tạo đường dẫn đầy đủ của tệp
        file_path = os.path.join(folder_path, file_name)

        # Trích xuất phần số và phần mở rộng từ tên tệp bằng biểu thức chính quy
        match = re.search(r'(\d+)_([a-zA-Z]+)\.([a-zA-Z]+)$', file_name)
        if match:
            # Lấy phần số từ tên tệp
            file_number = match.group(1)
            
            # Lấy phần mở rộng từ tên tệp
            file_extension = match.group(2)

            # Tạo tên mới cho tệp mask
            new_file_name = f"{file_number}.{file_extension}"

            # Đường dẫn đầy đủ của tệp mới
            new_file_path = os.path.join(folder_path, new_file_name)

            # Kiểm tra xem tệp mới đã tồn tại chưa
            if not os.path.exists(new_file_path):
                # Đổi tên tệp
                os.rename(file_path, new_file_path)
                print(f"Đã đổi tên '{file_name}' thành '{new_file_name}'")

# Sử dụng hàm để đổi tên các tệp trong thư mục
rename_files_in_folder('./Mask/valid/USSH')
```

```python
import os
import cv2
import numpy as np

def resize_image(image, target_size=(224, 224)):
    if image is None:
        return None
    
    if image.size == target_size:  # Kiểm tra xem hình ảnh đã ở kích thước mong muốn chưa
        return image
    
    resized_image = cv2.resize(image, target_size)
    return resized_image

def create_npy_from_folder(image_folder, mask_folder, output_folder):
    # Kiểm tra xem thư mục lưu trữ tệp `.npy` đã tồn tại chưa
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    image_data = []
    mask_data = []

    # Lặp qua các tệp trong thư mục hình ảnh và thư mục mặt nạ
    for image_file in os.listdir(image_folder):
        image_path = os.path.join(image_folder, image_file)
        mask_file = os.path.join(mask_folder, image_file)

        # Đảm bảo tệp mặt nạ tồn tại
        if os.path.exists(mask_file):
            # Đọc hình ảnh và mặt nạ
            image = cv2.imread(image_path, cv2.IMREAD_COLOR)
            mask = cv2.imread(mask_file, cv2.IMREAD_GRAYSCALE)

            # Resize hình ảnh thành kích thước 224x224x3
            resized_image = resize_image(image)
            if resized_image is None:
                continue  # Bỏ qua nếu hình ảnh rỗng

            # Chuyển đổi hình ảnh và mặt nạ thành mảng NumPy
            image_data.append(resized_image)
            mask_data.append(mask)

    # Chuyển danh sách các mảng thành một mảng NumPy
    image_arr = np.array(image_data)
    mask_arr = np.array(mask_data)

    # Lưu mảng vào tệp `.npy`
    np.save(os.path.join(output_folder, "image_arr.npy"), image_arr)
    np.save(os.path.join(output_folder, "mask_arr.npy"), mask_arr)

# Đường dẫn tới các thư mục chứa hình ảnh và mặt nạ
image_folder = "./data/"
mask_folder = "./Mask/"
output_folder = "./npy_data"

# Tạo các tệp `.npy` từ các thư mục hình ảnh và mặt nạ
create_npy_from_folder(image_folder, mask_folder, output_folder)
```

```python
import os
import glob
import numpy as np
import cv2

def resize_image(image, target_size=(224, 224)):
    resized_image = cv2.resize(image, target_size)
    return resized_image

def create_npy_from_folders(image_folder, mask_folder, output_folder):
    # Kiểm tra thư mục đầu ra tồn tại chưa, nếu không thì tạo mới
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Lấy danh sách các thư mục con trong thư mục hình ảnh
    subfolders = next(os.walk(image_folder))[1]

    for subfolder in subfolders:
        # Tạo thư mục đích cho mỗi loại hình ảnh
        output_subfolder = os.path.join(output_folder, subfolder)
        os.makedirs(output_subfolder, exist_ok=True)

        # Lấy danh sách các tệp hình ảnh trong thư mục con
        image_files = glob.glob(os.path.join(image_folder, subfolder, "*.*"))

        for image_file in image_files:
            # Tạo đường dẫn tương ứng đến tệp mặt nạ
            mask_file = image_file.replace(image_folder, mask_folder)

            # Đảm bảo tệp mặt nạ tồn tại
            if os.path.exists(mask_file):
                # Đọc hình ảnh và mặt nạ
                image = cv2.imread(image_file)
                mask = cv2.imread(mask_file)

                # Resize hình ảnh và mặt nạ thành kích thước mong muốn
                resized_image = resize_image(image)
                resized_mask = resize_image(mask)
                print(resized_image.shape)
                print(resized_mask.shape)

                # Lưu hình ảnh và mặt nạ đã được resize vào thư mục đích
                image_output_path = os.path.join(output_subfolder, os.path.basename(image_file))
                mask_output_path = os.path.join(output_subfolder, os.path.basename(mask_file))

                np.save(image_output_path, resized_image)
                np.save(mask_output_path, resized_mask)
# Đường dẫn tới các thư mục chứa hình ảnh và mặt nạ
image_folder = "./data"
mask_folder = "./mask"
output_folder = "./npy_data"

# Tạo các tệp `.npy` từ các thư mục hình ảnh và mặt nạ
create_npy_from_folders(image_folder, mask_folder, output_folder)
```

```python
def load_data(imagePath, labelPath):
    # Load images and labels
    images = np.load(imagePath)
    labels = np.load(labelPath)

    # Check if data is loaded correctly
    if len(images) != len(labels):
        raise ValueError("Number of images and labels do not match.")

    return images, labels

imagePath = './npy_data/image_arr.npy'
labelPath = './npy_data/mask_arr.npy'
load_data(imagePath, labelPath)
```

```python
import os
import cv2
import numpy as np
from os import listdir
from sklearn.preprocessing import LabelBinarizer
import pickle

raw_folder = "./data/train/"
def save_data(raw_folder=raw_folder):

    print("Bắt đầu xử lý ảnh...")

    pixels = []
    labels = []

    for folder in listdir(raw_folder):
        if folder!='.DS_Store':
            print("Folder=",folder)
            for file in listdir(raw_folder  + folder):
                if file!='.DS_Store':
                    print("File=", file)
                    pixels.append( cv2.resize(cv2.imread(raw_folder  + folder +"/" + file),dsize=(224,224)))
                    labels.append( folder)

    pixels = np.array(pixels)
    labels = np.array(labels)

    encoder = LabelBinarizer()
    labels = encoder.fit_transform(labels)
    print(labels)

    file = open('pix.data', 'wb')
    pickle.dump((pixels,labels), file)
    file.close()

    return


# Đường dẫn tới thư mục chứa dữ liệu raw

# Gọi hàm save_data để chuyển đổi và lưu dữ liệu
save_data(raw_folder)
```
